```python
import numpy as np
import matplotlib.pyplot as plt

def norm1(A: np.ndarray) -> float:
    maximum = 0
    n = len(A)
    for line in range(n):
        if sum(np.abs(A[line])) > maximum:
            maximum = sum(np.abs(A[line]))
    return maximum

# функция нахождения матрицы F в методе Ньютона
def find_F(w0: np.ndarray) -> np.ndarray:
    F1 = w0[0]**2 + w0[1]**2 - 1
    F2 = w0[1] - np.tan(w0[0])
    return np.array([F1, F2])

# функция нахождения матрицы Якоби
def Jacobi_matrix(w0: np.ndarray) -> np.ndarray:
    J = np.array([[2*w0[0], 2*w0[1]],
                  [-1/(np.cos(w0[0]))**2, 1]])
    return J

# метод Ньютона с ограничением итераций
def newton_method(initial_guess, max_iter=100, tol=1e-6):
    w0 = initial_guess.copy()

    for i in range(max_iter):
        try:
            F = find_F(w0)
            if np.linalg.norm(F) < tol:
                break

            J = Jacobi_matrix(w0)
            # Решаем линейную систему J * delta = -F
            delta = np.linalg.solve(J, -F)
            w0 = w0 + delta

            # Проверяем, чтобы значения оставались разумными
            if np.any(np.abs(w0) > 10):
                print("Значения вышли за допустимые пределы")
                return None

        except (OverflowError, ValueError, np.linalg.LinAlgError) as e:
            print(f"Ошибка на итерации {i}: {e}")
            return None

    return w0

# для (x1, y1) начальное приближение (x0, y0) = (0.7, 0.7)
print("Первое решение:")
solution1 = newton_method(np.array([0.7, 0.7]))
if solution1 is not None:
    print("Решение:", solution1)

print("\nВторое решение:")
# для (x2, y2) начальное приближение (x0, y0) = (-0.7, -0.7)
solution2 = newton_method(np.array([-0.7, -0.7]))
if solution2 is not None:
    print("Решение:", solution2)
```
